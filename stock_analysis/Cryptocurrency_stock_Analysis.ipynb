```python
bit['Open'].plot(label='bitcoin',figsize=(16,8),title='Open Price')
xrp['Open'].plot(label='XRP',color="green")
```

## Daily percentage change

```python
#bit['Volume'] = bit['Volume'].astype(float64)
bit['Volume'].plot(label='Bitcoin',figsize=(16,8),title='Volume Traded')
xrp['Volume'].plot(label='xrp',color="green")
```

```python
bit['returns'] = bit['Close'].pct_change(1)
```

```python
xrp['returns'] = xrp['Close'].pct_change(1)
```

```python
bit.head(3)
```

```python
xrp.head(3)
```

```python
bit['returns'].hist(bins=100,label='Bitcoin',figsize=(10,8),alpha=0.5)
xrp['returns'].hist(bins=100,label='XRP',alpha=0.5,color="Green")
```

```python
bit['returns'].plot(kind='kde',label='Bitcoin',figsize=(12,6))
xrp['returns'].plot(kind='kde',label='XRP',color="green")

```

```python
box_df = pd.concat([bit['returns'],xrp['returns'],],axis=1)
box_df.plot(kind='box',figsize=(8,11),colormap='jet')
```

## Cumulative Daily Returns

```python
bit['Cumulative Return'] = (1 + bit['returns']).cumprod()
```

```python
bit.head(3)
```

```python
xrp['Cumulative Return'] = (1 + xrp['returns']).cumprod()
```

```python
xrp.head(3)
```

```python
bit['Cumulative Return'].plot(label='Bitcoin',figsize=(16,8),title='Cumulative Return')
xrp['Cumulative Return'].plot(label='XRP',color="red")
```

```python
bit['Market Cap'].plot(label='Bitcoin',figsize=(16,8),title='Cumulative Return')
xrp['Market Cap'].plot(label='Xrp',color="green")
```
